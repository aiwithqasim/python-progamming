Q1. Write a program that accepts a list from user and print the alternate element of list.

```python
lst = []
while(True):
    user_input = input("enter anything(if you want to stop the input say 'y'): ")
    if(user_input == 'y'):
        break
    if(not(user_input.isalpha())):
        lst.append(int(user_input))
    else:    
        lst.append(user_input)    
print(lst)    

for x in range(0,len(lst),2):
    print(lst[x])
```

Q2. Write a program that accepts a list from user. Your program should reverse the content of list and 
display it. Do not use reverse() method.

```python
lst = []
while(True):
    user_input = input("enter anything(if you want to stop the input say 'y'): ")
    if(user_input == 'y'):
        break
    if(not(user_input.isalpha())):
        lst.append(int(user_input))
    else:    
        lst.append(user_input)    
print(lst) 

for x in lst:
    print(x[::-1])
```

Q8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be 
like this :   
- [2, 11, 7, 12]
- [5, 2, 9, 15]
- [8, 3, 10, 42]
- Sum of row 1 = 32 
- Sum of row 2 = 31 
- Sum of row 3 = 63



```python
lst = [[2,11,7,12],[5,2,9,15],[8,3,10,42]]
count = 1
for x in range(len(lst)):
    sum = 0
    for y in range(len(lst[x])):
        sum+=lst[x][y]
    print(f"Sum of row {count} = {sum}")    
    count+=1    
    


    
```
