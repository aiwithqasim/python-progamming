```python
class students():
    def __init__ (self, name, age):
        self.name = name
        self.age = age

    def details(self):
        print(f"{self.name} is of {self.age} age")
```

```python
stu1 = students('Qasim', '15')
```

```python
stu1.details()
```

```python
stu1.age
```

```python
for i in range(2): # external loop
    name = ''
    for student in ['Qasim', 'Ali']: # internal nested loop
        name = name + ' ' + student
    print(name)
    print('----')
```
