## Nested Loops

```python
first_names = ['Qasim', 'Ali', 'Hussain', 'Arham']
last_names = ['Hassan', 'Raza', 'Khan', 'Bhatti']

for f_name in first_names:
    for l_name in last_names:
        print(f"{f_name} {l_name}")
    print('----')
```

```python
lst1 = [[2,3,4,4],[2,3,4,5]] # 2 length

lst2 = [[6,7,8,9],[3,5,6,2]] # 2 length

result = []
for i in range(len(lst1)):
    # print(lst1[i])
    for j in range(len(lst1[0])):
        # print(lst1[i][j])
        result.append(lst1[i][j] + lst2[i][j])
result
```

## Dictionaries

- dictionaries are nothing but key-value pairs

```python
bio_data_list = ["Qasim", '5.6', 20, 'ABC address']

bio_data_dic = {
    'name' : 'Qasim',
    'height' : '5.6',
    'weight': 20,
    'address': 'ABC address'
}
```

```python
bio_data_dic
```

```python
bio_data_list[0] # extract value in list
bio_data_dic['name'] # extract value in dictionary
```

```python
bio_data_list.append('0900-7801') # add value in list
bio_data_dic['phone'] = '0900-78601' # add value in dictionary
```

```python
bio_data_dic 
```

```python
bio_data_dic.keys()
```

```python
bio_data_dic.values()
```

```python
bio_data_dic.items()
```

```python
# Apply for loop on dictionary bio_data_dic

for key in bio_data_dic.keys():
    print(key)
```

```python
for value in bio_data_dic.values():
    print(value)
```

```python
for data in bio_data_dic.items():
    print(data)
```

```python
for key, value in bio_data_dic.items():
    print(f"{key} --> {value}")
```
