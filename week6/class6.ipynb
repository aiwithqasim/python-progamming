## Create a list that contain dictionary

```python
customers = []
```

```python
customer0
```

```python
customer1 = {'customer id': 1,
 'first name': 'Qasim',
 'last name': 'Hassan',
 'address': '301 Stadium Rd.',
 'orders': []}
```

```python
customers.append(customer1)
```

```python
customers.append(customer0)
```

```python
customers
```

## How to get information out of a dictionary within a list

```python
customers
```

```python
# print all key value pairs that are in a dictionary with in a list

for customer in customers:
    for key, value in customer.items():
        print(f"{key} --> {value}")
    print('-----')
```

```python
## pirnt all details of student 1 and student 2

students = {
    "student1": {
        "name": "Alice",
        "age": 20,
        "grades": {
            "math": 90,
            "science": 85,
            "english": 88
        }
    },
    "student2": {
        "name": "Bob",
        "age": 22,
        "grades": {
            "math": 75,
            "science": 80,
            "english": 78
        }
    }
}
```

```python
for values in students.values():
    for key, val in values.items():
        print(f"{key} --> {val}")
    print('---')
```
