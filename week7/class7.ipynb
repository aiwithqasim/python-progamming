```python
## TASK: Write a function that calculate BMI of a person
## defualt argument/parameters

def bmi_cal(name, weight=20, height=5.1):
    return (f"{name} BMI is {weight/(height**2)}")
```

```python
bmi_cal('Qasim')
```

```python
bmi_cal("Arham", 50, 5.5)
```

```python
# SyntaxError: parameter without a default follows parameter with a default
# def bmi_cal(weight=20, height=5.1, name):
#    return (f"{name} BMI is {weight/(height**2)}")
```

```python
def bmi_cal(fname, lname, weight=20, height=5.1):
    return (f"{fname} {lname} BMI is {weight/(height**2)}")
```

```python
bmi_cal('Hassan', 'Qasim')
```

```python
bmi_cal(lname='Hassan', fname="Qasim") # key word arguments
```
