### Custom class
- Everything in python belongs to some **CLASS**.
- We create **OBJECTS** to interact with those classes.
- Every object/class can possess methods(functions).

NOTE: 
- If functions is created inside class it will be known as METHOD.
- Every METHOD and variable of a class will be handled via **SELF**

```python
class patient(): # class
    def patient_detail(self):
        print('This is a method of class Patient')


pid12 = patient() # object
pid12.patient_detail()
```

```python
class patient(): # class
    def patient_detail(self, name): # name is a argument of a methods of a class
        print('This is a method of class Patient')
        print(f'{name} is a patient')

pid12 = patient() # object
pid12.patient_detail("Ahram") # passing parameter to method of a class
```

```python
class patient(): # class
    def __init__(self, name): # constructor
        self.name = name # class variable initialization

    def patient_detail(self): # methods
        print('This is a method of class Patient')
        print(f'{self.name} is a patient')

pid12 = patient('Arham') # object, paramter in object
pid12.patient_detail()
```

NOTES:
- Every Class can have two things
    1. Characteristics/Properties (CONSTRUCTOR)
    2. Behaviour/Actions (METHODS)

- contructor will always be one while methods can be more than one.

```python
# TASK-2: Update the student class to have paramter from object initialization via constructor

class students(): #class

    def __init__(self, total_girls, total_boys): #constructor (Charactistics)
        self.teach_name = "Arfa" #hard coded the name here
        self.total_girls = total_girls # initializing class variables
        self.total_boys = total_boys
    
    def student_details(self): #method (Behavior/Action)
        print(f"We have {self.total_boys} boys and {self.total_girls} girls.")
        print(f"Teacher of the class is {self.teach_name}")

std_obj = students(15, 17) #obj, paramter in object i:e., 15, 17
std_obj.student_details() # method calling via obj
```

## Classes Topic Car Eample

- Characteristics/ Properties (Constructor: should be only one per class)
    - Type 
    - Model
    - Make (years)
    - color

- Behaviours/ Actions (Methods: can be more than one)
    - can travel
    - details

```python
class car():

    helpline = '0900-78601' # class variable

    def __init__(self, type, model, make, color): # constructor
        self.type = type # object variables
        self.model = model
        self.make = make
        self.color = color

    def can_travel(self): # method 1
        print(f"Automobiles (CARS) can help us in taveling")

    def details(self): # method 2
        print(f"--- Car Characterisics ----")
        print(f"TYPE is: {self.type}" )
        print(f"MODEL is: {self.model}" )
        print(f"MAKE is: {self.make}" )
        print(f"COLOR is: {self.color}" )

    def reset_color(self, color):
        self.color = color

obj_car = car("TOYOTA", "Corolla", 2026, "White")
```

```python
obj_car.reset_color('Black')
```

```python
obj_car.color
```

```python
obj_car.helpline # class variable
```

```python
## characteristics
print(obj_car.type)
print(obj_car.model)
print(obj_car.make)
print(obj_car.color)

## methods
print('------- METHODS -----------')

obj_car.can_travel()

obj_car.details()
```
