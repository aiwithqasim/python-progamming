## File Handling

```
with open(<file_name>, <mode>) as <alias>:
    <alias>.write("")
```

````python
with open("greeting.txt", "w") as f:
    f.write("Hello, World!")
````

````python
with open("greeting.txt", "r") as f:
    content = f.read()

print(content)
````

````python
with open("greeting.txt", "a") as f:
    f.write(". Welcome to Python programming!")
````

## CSV Files
- read(r) -> used for reading a file
- write(w) -> used for writing to a file (overwrites existing content)
- append(a) -> used for appending to a file

````python
import csv

with open("competitions.csv", "r") as f:
    content = csv.reader(f)
    empty_list = []
    for row in content:
        empty_list.append(row)
print(empty_list)
````

````python
empty_list[0]
````

````python
empty_list[1].index("Best-Kept Lawn")
````

````python
with open("competitions.csv", "a", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["2026", "Saylani Hackathon", "Arham"])
````
